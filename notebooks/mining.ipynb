```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# =========================
# 1. Prepare data
# =========================
prod = product_in_city_df.copy()

prod['label'] = prod['product_name'] + '\n(' + prod['category'] + ')'

# color by category
categories = prod['category'].unique()
palette = dict(zip(categories, ['#d62728', '#2ca02c','#1f77b4','#ff7f0e','#9467bd']))
prod['color'] = prod['category'].map(palette)

# size by total orders
prod['size'] = (prod['total_orders'] / prod['total_orders'].max() * 300).clip(lower=40)

# city order
city_list = ['Vadodara', 'Surat', 'Ahmedabad']

# global axis range for consistency
x_min = prod['late_pct'].min() - 1
x_max = prod['late_pct'].max() + 1
y_min = prod['fill_rate'].min() - 0.2
y_max = prod['fill_rate'].max() + 0.2

# =========================
# 2. Create subplots
# =========================
fig, axes = plt.subplots(3, 1, figsize=(10, 20), sharex=True, sharey=True)

for ax, city in zip(axes, city_list):
    city_df = prod[prod['city'] == city].copy()
    
    # median per city
    med_fill = city_df['fill_rate'].median()
    med_late = city_df['late_pct'].median()
    
    # worst products in each city
    worst_prod = city_df[
        (city_df['late_pct'] > med_late) &
        (city_df['fill_rate'] < med_fill)
    ]
    
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)

    # quadrant shading
    y_norm = (med_fill - y_min) / (y_max - y_min)

    ax.axvspan(x_min, med_late, ymin=0,      ymax=y_norm, alpha=0.06, color='orange')
    ax.axvspan(med_late, x_max, ymin=0,      ymax=y_norm, alpha=0.10, color='red')
    ax.axvspan(x_min, med_late, ymin=y_norm, ymax=1,      alpha=0.06, color='green')
    ax.axvspan(med_late, x_max, ymin=y_norm, ymax=1,      alpha=0.04, color='orange')

    # scatter
    ax.scatter(
        city_df['late_pct'],
        city_df['fill_rate'],
        c=city_df['color'],
        s=city_df['size'],
        alpha=0.85,
        edgecolor='black',
        linewidth=0.4,
        zorder=3
    )

    # median lines
    ax.axvline(x=med_late, color='gray', linewidth=0.8, linestyle='--')
    ax.axhline(y=med_fill, color='gray', linewidth=0.8, linestyle='--')

    # titles and labels
    ax.set_title(city, fontsize=12, weight='bold')
    ax.set_xlabel('% Orders Late')

axes[0].set_ylabel('Fill Rate %')

# =========================
# 3. Legend
# =========================
color_legend = [
    Line2D([0], [0], marker='o', color='w',
           markerfacecolor=palette[cat], markeredgecolor='black',
           markersize=9, label=cat)
    for cat in categories
]

fig.legend(
    handles=color_legend,
    loc='lower center',
    ncol=len(categories),
    fontsize=9,
    frameon=False,
    title='Category'
)

fig.suptitle('Product Performance by City — Late % vs Fill Rate', fontsize=15, weight='bold')
plt.tight_layout(rect=[0, 0.08, 1, 0.95])
plt.show()
```

```python
query = """
SELECT 
    c.city,
    o.product_id,
    p.product_name,
    p.category,
    CAST(SUM(o.actual_revenue) AS NUMERIC) as actual_revenue,
    CAST(SUM(o.expected_revenue) AS NUMERIC) as expected_revenue,
    ROUND(CAST(SUM(GREATEST(o.expected_revenue - o.actual_revenue, 0)) AS NUMERIC), 0) AS lost_revenue,
    ROUND(AVG(SAFE_DIVIDE(o.delivered_qty, o.order_qty)) * 100, 2) AS avg_fill_rate,
    ROUND(COUNTIF(o.actual_delivery_date > o.agreed_delivery_date) / COUNT(o.order_id) * 100, 2) AS avg_late_pct
FROM analytics.dim_customers c
JOIN analytics.fact_order_lines o ON c.customer_id = o.customer_id
JOIN analytics.dim_products p ON p.product_id = o.product_id
GROUP BY c.city, o.product_id, p.product_name, p.category
"""
perf_in_city_df = client.query(query).to_dataframe()
perf_in_city_df
```

```python


# 1. Gom nhóm theo City và Category
# Chúng ta dùng mean cho các tỉ lệ % và sum cho doanh thu mất đi
summary_df = perf_in_city_df.groupby(['city', 'category']).agg({
    'avg_fill_rate': 'mean',
    'avg_late_pct': 'mean',
    'lost_revenue': 'sum'
}).reset_index()

summary_df = summary_df.sort_values(by='city', ascending=False)

summary_df['avg_fill_rate'] = summary_df['avg_fill_rate'].round(2)
summary_df['avg_late_pct'] = summary_df['avg_late_pct'].round(2)

final_report = summary_df[['city', 'category', 'avg_fill_rate', 'avg_late_pct', 'lost_revenue']]

print(final_report)
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np


# 1. Tạo bản sao để tránh ảnh hưởng df gốc
df_model = perf_in_city_df.copy()
df_model['lost_rev_pct'] = (df_model['lost_revenue'] / (df_model['expected_revenue'])) * 100

# 2. Ép kiểu 3 cột chính về float
cols_to_fix = ['lost_rev_pct', 'avg_fill_rate', 'avg_late_pct']
for col in cols_to_fix:
    df_model[col] = pd.to_numeric(df_model[col], errors='coerce')

# 3. Tạo biến giả (Dummies)
df_model = pd.get_dummies(df_model, columns=['category', 'city'], drop_first=True)

# 4. CHỐT CHẶN CUỐI: Ép toàn bộ dataframe về float64 để loại bỏ kiểu 'object'
df_model = df_model.select_dtypes(include=[np.number, 'bool']).astype(float)

# 5. Xác định X và y
# Lưu ý: Tên cột dummies có thể thay đổi tùy vào dữ liệu của bạn, 
# hãy kiểm tra df_model.columns để điền chính xác
X_cols = [col for col in df_model.columns if col != 'lost_rev_pct']
X = df_model[X_cols]
X = sm.add_constant(X)
y = df_model['lost_rev_pct']

# 6. Chạy mô hình
model = sm.OLS(y, X).fit()

# 7. Xem kết quả
print(model.summary())
```

```python
%%sql delay_df<<
SELECT
    f.customer_id,
    c.customer_name,
    c.city,
    ROUND(AVG(
        DATE_DIFF(f.actual_delivery_date, f.agreed_delivery_date, DAY)
    ), 2) AS avg_delay_days
FROM analytics.fact_order_lines f
JOIN analytics.dim_customers c ON f.customer_sk = c.customer_sk
WHERE f.actual_delivery_date > f.agreed_delivery_date
GROUP BY f.customer_id, c.customer_name, c.city
ORDER BY avg_delay_days DESC
```

```python
query = """
WITH order_sizes AS (
    -- Phân loại Size cho từng Order
    SELECT 
        order_id,
        customer_id,
        NTILE(4) OVER (ORDER BY SUM(order_qty)) AS size_group_num
    FROM analytics.fact_order_lines
    GROUP BY order_id, customer_id
),
labeled_orders AS (
    SELECT 
        os.order_id,
        os.customer_id,
        CASE 
            WHEN size_group_num = 1 THEN 'Small'
            WHEN size_group_num = 2 THEN 'Medium'
            WHEN size_group_num = 3 THEN 'Large'
            WHEN size_group_num = 4 THEN 'Very Large'
        END AS order_size_group
    FROM order_sizes os
)
-- Tính tỷ trọng đơn Small của từng khách hàng
SELECT 
    c.customer_name,
    COUNT(lo.order_id) AS total_orders,
    COUNTIF(lo.order_size_group = 'Small') AS small_orders,
    ROUND(COUNTIF(lo.order_size_group = 'Small') / COUNT(lo.order_id) * 100, 2) AS small_pct
FROM labeled_orders lo
JOIN analytics.dim_customers c ON lo.customer_id = c.customer_id
GROUP BY c.customer_name
ORDER BY small_pct DESC
"""
customer_behavior_df = client.query(query).to_dataframe()
customer_behavior_df.head(10)
```

```python
import plotly.express as px

# Cắt top 10 khách hàng có tỷ lệ đơn Small cao nhất
top_small_customers = customer_behavior_df.head(10)

fig = px.bar(
    top_small_customers,
    x='small_pct',
    y='customer_name',
    orientation='h',
    title='Pecentage of Small Oders by Customer',
    text='small_pct',
    labels={'small_pct': '% Small Orders', 'customer_name': 'Customer name'}
)
fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.show()
```
